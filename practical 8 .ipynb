# Practical 8:	Use the iris dataset of the sklearn library to create the following visual representations of data using matplotlib and/or seaborn library:
* a.	Scatter plot showing the relationship between petal length and petal width of different instances of iris flowers. 
* b.	Histograms showing the data distribution of each of the four attributes.
* c.	Pie Chart showing the frequency count of each flower type.
* d.	Pair Plot showing the relationship between every pair of attributes 

#  a.	Scatter plot showing the relationship between petal length and petal width of different instances of iris flowers. 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset from seaborn
iris = sns.load_dataset("iris")

# Scatter plot for petal length vs petal width
sns.scatterplot(x="petal_length", y="petal_width", data=iris, hue="species")

# Set plot title and labels
plt.title("Scatter Plot of Petal Length vs Petal Width")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")

# Show the plot
plt.show()
```

# * b.	Histograms showing the data distribution of each of the four attributes.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset from seaborn
iris = sns.load_dataset("iris")

# Create histograms for each attribute
sns.histplot(data=iris, x="sepal_length", hue="species", kde=True, multiple="stack", bins=15)
plt.title("Histogram of Sepal Length by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()

sns.histplot(data=iris, x="sepal_width", hue="species", kde=True, multiple="stack", bins=15)
plt.title("Histogram of Sepal Width by Species")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frequency")
plt.show()

sns.histplot(data=iris, x="petal_length", hue="species", kde=True, multiple="stack", bins=15)
plt.title("Histogram of Petal Length by Species")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.show()

sns.histplot(data=iris, x="petal_width", hue="species", kde=True, multiple="stack", bins=15)
plt.title("Histogram of Petal Width by Species")
plt.xlabel("Petal Width (cm)")
plt.ylabel("Frequency")
plt.show()
```

# c.	Pie Chart showing the frequency count of each flower type.

```python
import matplotlib.pyplot as plt

species_count = iris['species'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(species_count, labels=species_count.index, autopct='%1.1f%%')
plt.title('Frequency Count of Each Flower Type')
plt.show()
```

# d.	Pair Plot showing the relationship between every pair of attributes 

```python
sns.pairplot(iris, hue='species')
plt.show()
```
